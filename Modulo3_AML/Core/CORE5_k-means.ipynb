```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Selección de variables
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Implementación de K-means para diferentes valores de K
k_values = [3, 4, 5]
for k in k_values:
  kmeans = KMeans(n_clusters=k, random_state=0)
  kmeans.fit(X)
  labels = kmeans.labels_
  centroids = kmeans.cluster_centers_

  # Visualización de los resultados
  plt.figure(figsize=(8, 6))
  plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'], c=labels, s=50, cmap='viridis')
  plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='*', label='Centroids')
  plt.xlabel('Annual Income (k$)')
  plt.ylabel('Spending Score (1-100)')
  plt.title(f'K-means Clustering (k={k})')
  plt.legend()
  plt.show()
```

```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

drive.mount('/content/drive')

ruta_archivo = '/content/drive/My Drive/Colab Notebooks/Dataset/Mall_Customers.csv'

df = pd.read_csv(ruta_archivo)

# Selección de variables
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Implementación del método del codo
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Visualización del método del codo
plt.figure(figsize=(10,5))
plt.plot(range(1, 11), wcss)
plt.title('Método del Codo')
plt.xlabel('Número de Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Ejemplo con k=3 (adaptar según el análisis del método del codo)
kmeans = KMeans(n_clusters=3, random_state=0) # Reemplazar 3 por el k óptimo
kmeans.fit(X)
df['Cluster'] = kmeans.labels_

# Análisis descriptivo de cada cluster
for cluster in range(3): # Iterar sobre el numero de clusters
    cluster_data = df[df['Cluster'] == cluster]
    print(f"Análisis del Cluster {cluster}:\n")
    print(cluster_data[['Annual Income (k$)', 'Spending Score (1-100)']].describe())
    print("-" * 30)


# Estrategias de marketing personalizadas (ejemplo)
print("Estrategias de Marketing:\n")

# Cluster 0: Supongamos que este cluster tiene ingresos medios-altos y un gasto medio
print("Cluster 0 (Ingresos medios-altos, gasto medio):")
print("  - Ofrecer productos de gama media-alta con promociones exclusivas.")
print("  - Programas de fidelización con beneficios atractivos.")
print("  - Campañas publicitarias en medios digitales enfocados a su estilo de vida.\n")

# Cluster 1: Supongamos que este cluster tiene ingresos bajos y un gasto bajo
print("Cluster 1 (Ingresos bajos, gasto bajo):")
print("  - Promociones y descuentos agresivos para incentivar la compra.")
print("  - Productos de precio accesible.")
print("  - Publicidad en medios con mayor alcance masivo.\n")

# Cluster 2: Supongamos que este cluster tiene ingresos altos y un gasto alto
print("Cluster 2 (Ingresos altos, gasto alto):")
print("  - Productos premium y exclusivos.")
print("  - Experiencias de compra personalizadas.")
print("  - Atención personalizada y asesoramiento especializado.\n")

#Conclusión (adaptar a los resultados obtenidos)

print("Conclusión:")
print("La segmentación de clientes a través del análisis de clusters nos proporciona información valiosa para la creación de estrategias de marketing más efectivas. Al comprender las características de cada grupo, podemos dirigir las campañas y las ofertas a las necesidades específicas de cada cliente, maximizando la conversión y la retención.")
```
