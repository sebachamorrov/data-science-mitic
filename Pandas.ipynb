```python
# Inciso de funciones. Una función es un conjunto de instrucciones encapsuladas que pueden
# ser llamadas en cualquier momento (como una caja negra)
def suma(a, b):
    return a + b

x = suma(1, 2)
print("El resultado es", x)


def op_mat(a, b):
    x1 = a + b
    x2 = a - b
    x3 = a * b
    x4 = a/b
    return x1, x2, x3, x4

print("Resultado de operaciones matemáticas:",op_mat(1, 2))
```

```python
# Creamos la función pow2 que calcula el cuadrado de un número
def pow2(n): # Función que retorna n elevado a 2
    return n**2

# Ejemplo de uso
pow2(4) # Resultado 16
# Aplicaremos esta función pow2 a todos los elementos de
# nuestro dataframe. Para ello, entregamos la función
# pow2 como entrada a .apply
df['A'].apply(pow2)
```

```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {
    'Producto': ['Manzanas', 'Naranjas', 'Bananas', 'Peras'],
    'Ventas': [50, 30, 20, 60],
    'Precio': [0.5, 0.75, 0.3, 0.4]
}
df = pd.DataFrame(data)


# Filtrar filas donde las ventas son mayores a 25
ventas_mayores_25 = df[df['Ventas'] > 25]
print(ventas_mayores_25)

# Filtrar filas donde las ventas son mayores a 25 y precio menor que 0.5
ventas_mayores_25_menores_05 = df[(df['Ventas'] > 25) & (df['Precio'] < 0.5)]
print(ventas_mayores_25_menores_05)

# Filtrar filas utilizando el método query
query_result = df.query('Ventas > 25 & Precio < 0.5')
print(query_result)

# Filtrar filas utilizando el índice
df_index = df.set_index('Producto')
df_index.head()
df_index = df.set_index('Producto')
ventas_manzanas = df_index.loc['Manzanas']
print(ventas_manzanas)

# Filtrar filas donde el producto es 'Manzanas' o 'Bananas'
ventas_manzanas_bananas = df[df['Producto'].isin(['Manzanas', 'Bananas'])]
print(ventas_manzanas_bananas)

loc_result = df.loc[0:2, ['Producto', 'Ventas']]
print(loc_result)

iloc_result = df.iloc[0:3, 0:2]
print(iloc_result)

# Filtrar y luego hacer slicing de columnas
filtered_sliced = df[df['Ventas'] > 25].loc[:, ['Producto', 'Ventas']]
print(filtered_sliced)
```
