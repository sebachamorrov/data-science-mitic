```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
df
```

```python
df['InvoiceDate'] = df['InvoiceDate'].dt.date
df
```

```python
df['UnitPrice'].sum()
```

```python
df.isnull().sum()
```

```python
df_null = df[df['Description'].isnull()]
df_null
```

```python
df_null['UnitPrice'].sum()
```

```python
df['Venta'] = df.apply(lambda row: row['Quantity'] * row['UnitPrice'], axis=1)
df
```

```python
df['Tipo_Doc'] = df['Venta'].apply(lambda x: 'Nota de Credito' if x <= 0 else 'Factura')
df
```

```python
vta_max = df['Venta'].max()
vta_min = df['Venta'].min()
df['Vta_Normalizada'] = df['Venta'].apply(lambda x: (x - vta_min) / (vta_max - vta_min))
df
```

```python
venta_por_pais = df.groupby('Country')['Venta'].sum().sort_values(ascending=False)
venta_por_pais.head()
```

```python
venta_media_por_pais = df.groupby('Country')['Venta'].mean().sort_values(ascending=False)
venta_media_por_pais.head()
```

```python
venta_cant_por_pais = df.groupby('Country')['Venta'].count().sort_values(ascending=False)
venta_cant_por_pais.head()
```

```python
venta_max_por_pais = df.groupby('Country')['Venta'].max().sort_values(ascending=False)
venta_max_por_pais.head()
```

```python
venta_min_por_pais = df.groupby('Country')['Venta'].min().sort_values(ascending=False)
venta_min_por_pais.head()
```

```python
venta_desv_por_pais = df.groupby('Country')['Venta'].std().sort_values(ascending=False)
venta_desv_por_pais.head()
```

```python
venta_var_por_pais = df.groupby('Country')['Venta'].var().sort_values(ascending=False)
venta_var_por_pais.head()
```
