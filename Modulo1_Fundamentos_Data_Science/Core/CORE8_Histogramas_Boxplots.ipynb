```python
import kagglehub
import os
import pandas as pd
import matplotlib.pyplot as plt

# Descargar la última versión del dataset
path = kagglehub.dataset_download("unsdsn/world-happiness")

# Definir la ruta a la carpeta que contiene los archivos CSV
dataset_path = path  # Esto ya apunta a la carpeta 'versions/2'

# Listar y cargar archivos CSV en un solo DataFrame
csv_files = [file for file in os.listdir(dataset_path) if file.endswith('.csv')]

if csv_files:
    # Cargar todos los archivos CSV en un solo DataFrame
    df_all = pd.concat([pd.read_csv(os.path.join(dataset_path, file)) for file in csv_files], ignore_index=True)
    
    # Ver las primeras filas del DataFrame combinado
    print(df_all.head())
else:
    print("No se encontraron archivos CSV para cargar.")
```

##### Histograma

```python
# Histograma
datos = df['Happiness.Score']  # Se cargan los datos

plt.hist(datos, bins=5, rwidth=0.5)  # bins define el número de intervalos, rwidth ajusta el ancho de las barras

# Etiquetas y título
plt.xlabel("Happines Score")  # Etiqueta del eje X
plt.ylabel("Frecuencia")      # Etiqueta del eje Y
plt.title("Histograma Happines Score")  # Título del gráfico

# Mostrar gráfico
plt.show()
```

El happines score mundial se mantiene principalmente entre los rangos de 5 y 6

```python
# Crear subgráficas
fig, axs = plt.subplots(1, 2, figsize=(12, 5))  # 1 fila, 2 columnas

# Histograma
datos = df['Freedom']  # Se cargan los datos
n, bins, patches = axs[0].hist(datos, bins=5, rwidth=0.5)  # Histograma
axs[0].set_xlabel("Freedom")  # Etiqueta del eje X
axs[0].set_ylabel("Frecuencia")  # Etiqueta del eje Y
axs[0].set_title("Histograma Freedom")  # Título del gráfico

# Etiquetas en el histograma
for count, x in zip(n, bins):
    axs[0].text(x + 0.05, count, str(int(count)), ha='center', va='bottom')

# Boxplot
axs[1].boxplot(datos)  # Boxplot
axs[1].set_ylabel("Freedom")  # Etiqueta del eje Y
axs[1].set_title("Boxplot Freedom")  # Título del gráfico

# Calcular los cuartiles
Q1 = datos.quantile(0.25)  # Primer cuartil
Q2 = datos.median()  # Mediana (segundo cuartil)
Q3 = datos.quantile(0.75)  # Tercer cuartil

# Etiquetas de datos para los cuartiles
axs[1].text(1.1, Q1, f'Q1: {Q1:.2f}', fontsize=10, ha='left')
axs[1].text(1.1, Q2, f'Median: {Q2:.2f}', fontsize=10, ha='left')
axs[1].text(1.1, Q3, f'Q3: {Q3:.2f}', fontsize=10, ha='left')

# Ajustar el espacio entre gráficos
plt.tight_layout()

# Mostrar gráfico
plt.show()
```
