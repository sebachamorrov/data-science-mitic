# RECREACION GRAFICO

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = sns.load_dataset('tips')
df
```

```python
# Crear una figura y un eje
fig, ax = plt.subplots(figsize=(12, 6))

# GrÃ¡fico de dispersiÃ³n con cuadrÃ­cula, leyenda, anotaciones y flechas
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex', style='smoker', s=50, ax=ax)
ax.set_title('RelaciÃ³n entre Total Bill y Tip con Anotaciones y Flechas', fontsize=20)
ax.set_xlabel('Total Bill ($)', fontsize=15)
ax.set_ylabel('Tip ($)', fontsize=15)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize='large', shadow=True, title='Detalles', framealpha=0.5, frameon=False)


# AÃ±adir cuadrÃ­cula
ax.grid(True, which='major', linestyle='--', linewidth=2, color='lightblue')

max_tip = df['tip'].max()
max_tip_row = df[df['tip'] == max_tip].iloc[0]
xy=(max_tip_row['total_bill'], max_tip_row['tip'])

# AÃ±adir anotaciÃ³n
ax.annotate('Mayor tip en la cena', xy=(50,10), xytext=(55,15),
            arrowprops=dict(facecolor='black', shrink=0.05))

# AÃ±adir flecha
ax.arrow(20, 5, 5, 3, head_width=0.8, head_length=0.5, fc='green', ec='green')

# Mostrar el grÃ¡fico
plt.show()
```
