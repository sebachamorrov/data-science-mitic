# Arreglar el gr√°fico (Core)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos de ejemplo en formato CSV
data = """
fecha,ventas
2023-01-01,200
2023-01-02,220
2023-01-03,210
2023-01-04,205
2023-01-05,215
2023-01-06,225
"""

data = {
    'Fecha': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05', '2023-01-06'],
    'Ventas': [200, 220, 210, 205, 215, 225]
}

print(data)
```

```python
# Cargar los datos en un DataFrame de Pandas
df = pd.DataFrame(data)
df
```

```python
# Validar tipos de datos
df.dtypes
```

```python
# Convertir la columna 'fecha' a formato datetime
df['Fecha'] = pd.to_datetime(df['Fecha'], errors='coerce')
df
```

```python
# Validar el cambio de tipo de dato
df.dtypes
```

```python
# Gr√°fico de l√≠neas de las ventas
plt.figure(figsize=(7, 3))
plt.plot(df['Fecha'], df['Ventas'])
plt.title("Ventas Diarias")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.show()
```

```python
# Crear un gr√°fico de dispersi√≥n
plt.figure(figsize=(7, 3))
plt.scatter(df['Fecha'], df['Ventas'])
plt.title("Ventas Diarias")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.show()
```

```python
# Crear un histograma de las ventas
plt.hist(df['Ventas'], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribuci√≥n de Ventas")
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.show()
```
